_**Why do we use Balanced Accuracy as a metric of evaluation?**_

==> The significance of balanced accuracy lies in its ability to provide a more accurate evaluation of a classifier's performance when dealing with imbalanced datasets. In imbalanced datasets, where one class is significantly more prevalent than the others, accuracy alone can be misleading. For instance, if a classifier simply predicts the majority class for all instances, it may achieve high accuracy, but it fails to capture the nuances of the other classes. Balanced accuracy addresses this issue by considering performance across all classes, providing a more balanced view of the classifier's effectiveness.

#### Balancing number of obs. for non-PE polymers by generating synthetic data

- **Reference**: https://pubs.acs.org/doi/full/10.1021/acs.estlett.2c00949

```python
# Filter out "PE" samples
non_pe_data = merged_df[merged_df['polymer'] != 'PE']

# Function to generate synthetic samples for a specific polymer class
def generate_synthetic_samples(inputdata, polymer_class, num_samples):
    polymer_data = inputdata[inputdata['polymer'] == polymer_class]
    synthetic_samples = []
    
    for _ in range(num_samples):
        # Pick two random samples
        sample_indices = np.random.choice(polymer_data.index, size=2, replace=False)
        V1 = polymer_data.loc[sample_indices[0], '976':'1800'].values
        V2 = polymer_data.loc[sample_indices[1], '976':'1800'].values
        
        # Generate two random weights that sum to 1
        w1, w2 = np.random.rand(2)
        w1, w2 = w1 / (w1 + w2), w2 / (w1 + w2)
        
        # Compute the new synthetic sample
        synthetic_sample = w1 * V1 + w2 * V2
        synthetic_samples.append(synthetic_sample)
    
    return synthetic_samples

# Define the number of synthetic samples to generate for each polymer class
num_synthetic_samples_per_class = 305 - non_pe_data['polymer'].value_counts()

# Generate synthetic samples for each non-PE polymer class
synthetic_samples = []

for polymer_class in num_synthetic_samples_per_class.index:
    samples = generate_synthetic_samples(inputdata = non_pe_data, polymer_class = polymer_class, num_samples = num_synthetic_samples_per_class[polymer_class])
    for sample in samples:
        sample_dict = {'polymer': polymer_class}
        sample_dict.update(dict(zip(merged_df.columns[1:], sample))) # update the synthetic sample list with wavenumber 975 to 1800
        synthetic_samples.append(sample_dict)

# Convert synthetic samples to a DataFrame
synthetic_df = pd.DataFrame(synthetic_samples)
```

## Data Normalization

### 1. (Total Sum Normalization) **Normalize by the sum of all wavelengths for each sample**

```python
def total_sum_normalization(df):
    # Calculate the sum of each row
    sum_per_sample = df.sum(axis=1)
    # Reshape the sum to match the original DataFrame's dimensions for broadcasting
    sum_per_sample = sum_per_sample.values.reshape(-1, 1)
    # Normalize each row by its sum
    normalized_df = df.div(sum_per_sample)
    return normalized_df
```

```python
tsn_spectra = total_sum_normalization(synthetic_data.drop(columns = ['polymer'])).reset_index(drop=True)
# Concatenate the original 'polymer' column to normalized spectra
synthetic_tsn_spectra = pd.concat([synthetic_data[['polymer']], tsn_spectra], axis=1)
```

### 2. Min-Max Scaler

```python
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the spectra, dropping the non-numerical columns and setting 'polymer_day' as index
spectra = synthetic_data.drop(columns=['polymer']).T
minmax_spectra = scaler.fit_transform(spectra).T

# Convert the numpy array back to a DataFrame
minmax_spectra_df = pd.DataFrame(minmax_spectra, columns=spectra.index, index=spectra.columns)

# Concatenate the original columns with the scaled spectra
synthetic_min_max_scaler_spectra = pd.concat([synthetic_data[['polymer']], minmax_spectra_df], axis=1)
```

### 3. (Z-score normalization) **The values were standardized to their mean value and unit variance for each wave number**

```python
def z_score_normalization(spectra):
    # Calculate the mean and standard deviation for each wave number (wave numbers as rows of input spectra df)
    means = np.mean(spectra, axis=0)
    stds = np.std(spectra, axis=0)
    
    # Perform Z-score normalization
    normalized_spectra = (spectra - means) / stds
    return normalized_spectra
```

```python
spectra = synthetic_data.drop(columns=['polymer']).T

# Apply the Z-score normalization
zscore_spectra = z_score_normalization(spectra).T

# Convert the numpy array back to a DataFrame
zscore_spectra_df = pd.DataFrame(zscore_spectra, columns=spectra.index, index=spectra.columns).reset_index(drop=True)

# Concatenate the original columns with the normalized spectra
synthetic_zscore_spectra = pd.concat([synthetic_data[['polymer']], zscore_spectra_df], axis=1)
```

### 4. (Log10-normalization of raw signal abundance)

```python
spectra = synthetic_data.drop(columns=['polymer']).T

log_spectra = np.log10(spectra).T

# Convert the numpy array back to a DataFrame
log_spectra_df = pd.DataFrame(log_spectra, columns=spectra.index, index=spectra.columns).reset_index(drop=True)

# Concatenate the original columns with the normalized spectra
synthetic_log_spectra = pd.concat([synthetic_data[['polymer']], log_spectra_df], axis=1)
```

### 5. Scale raw data twice (TSN then log10 normalization)

```python
spectra = synthetic_tsn_spectra.drop(columns=['polymer']).T

tsn_log_spectra = np.log10(spectra).T

# Convert the numpy array back to a DataFrame
tsn_log_spectra_df = pd.DataFrame(tsn_log_spectra, columns=spectra.index, index=spectra.columns).reset_index(drop=True)

# Concatenate the original columns with the normalized spectra
synthetic_tsn_log_spectra = pd.concat([synthetic_tsn_spectra[['polymer']], tsn_log_spectra_df], axis=1)
```

### 6. Data Enrichment with added AVG/STD./VAR. columns

```python
id_list = common_columns.tolist()
if 'polymer' in id_list:
    id_list.remove('polymer')
    
id_list = pd.Index(id_list)
first_deri_df = pd.concat([ori_dat[id_list], tmu_polymer[['sample']]], axis=1).set_index('sample').T

column_names = first_deri_df.columns.tolist()
split_names = [name.split("_")[0] for name in column_names]

string_list = {}

for name in set(split_names):
    string_list[name] = [i for i, val in enumerate(split_names) if val == name]

new_columns = []

for key, value in string_list.items():
    # Calculate row-wise derivatives (averages, var & std.) of all replicates of each polymer / each day
    avg_col = first_deri_df.iloc[:, value].mean(axis=1)
    std_col = first_deri_df.iloc[:, value].std(axis=1)
    var_col = first_deri_df.iloc[:, value].var(axis=1)
    
    # Append these new columns to the list
    new_columns.append(pd.DataFrame({
        f"AVG_{key}": avg_col,
        f"STD_{key}": std_col,
        f"VAR_{key}": var_col
    }))

# Concatenate all new columns at once
first_deri_df = pd.concat(new_columns, axis=1)

first_deri_df1 = first_deri_df.T
```

```python
# Function to determine the polymer type
def determine_polymer(sample):
    if re.search("ULDPE|LLDPE1|LLDPE2|LDPE1|LDPE2|MDPE|HDPE1|HDPE2", sample, re.IGNORECASE):
        return "PE"
    elif re.search("PP", sample, re.IGNORECASE):
        return "PP"
    elif re.search("PEST", sample, re.IGNORECASE):
        return "PEST"
    elif re.search("PET1|PET2", sample, re.IGNORECASE):
        return "PET"
    elif re.search("EVA", sample, re.IGNORECASE):
        return "EVA"
    elif re.search("ABS", sample, re.IGNORECASE):
        return "ABS"
    elif re.search("EPS", sample, re.IGNORECASE):
        return "EPS"
    elif re.search("PS", sample, re.IGNORECASE):
        return "PS"
    elif re.search("PA66|PA6", sample, re.IGNORECASE):
        return "PA"
    elif re.search("PVC1|PVC2", sample, re.IGNORECASE):
        return "PVC"
    elif re.search("CR", sample, re.IGNORECASE):
        return "CR"
    else:
        return "CA"
```

#### Extract Features from Images

```python
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16, preprocess_input
from keras.models import Model
import numpy as np
import os

# Load the VGG16 model pre-trained on ImageNet dataset
base_model = VGG16(weights='imagenet')
model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc1').output)

def extract_features(img_path, model):
    img = image.load_img(img_path, target_size=(224, 224))
    img_data = image.img_to_array(img)
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    
    features = model.predict(img_data)
    return features

# Extract features for all images and store them in a list
image_features = []
image_labels = []

for polymer in unique_polymers:
    image_path = os.path.join(output_dir, f'{polymer}_spectra_polar.png')
    features = extract_features(image_path, model)
    image_features.append(features)
    image_labels.append(polymer)

image_features = np.array(image_features)
image_labels = np.array(image_labels)
```

#### Train KNN Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Reshape image_features to the correct shape
image_features = image_features.reshape((len(image_features), -1))

# Encode the labels
le = LabelEncoder()
image_labels_encoded = le.fit_transform(image_labels)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(image_features, image_labels_encoded, test_size=0.2, random_state=42)

# Define and train the Random Subspace KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
classifier = BaggingClassifier(base_estimator=knn, n_estimators=10, max_features=0.5)
classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = classifier.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
# Run multiple times and collect accuracy scores
for i in range(num_iterations):
    # Randomly split the training set for more robust training
    X_tr_split1, X_tr_split2, y_tr_split1, y_tr_split2 = train_test_split(X_train, y_train, stratify = y_train, test_size=0.2, 
                                                                      random_state=random.randint(1, 999))
    
    # Randomly split the hold-out test set for more robust testing results
    X_t_split1, X_t_split2, y_t_split1, y_t_split2 = train_test_split(X_test, y_test, stratify = y_test, test_size=0.2, 
                                                                      random_state=random.randint(1, 999))
    
    # Encode the labels for bothn training and testing sets
    label_encoder = LabelEncoder()
    y_tr_split1 = label_encoder.fit_transform(y_tr_split1)
    y_t_split1 = label_encoder.fit_transform(y_t_split1)
    
    # Split features and labels
    X_tr = X_tr_split1.values
    y_tr = y_tr_split1
    X_t = X_t_split1.values
    y_t = y_t_split1
        
    # Define the model
    model = Sequential([
    tf.keras.layers.Input(shape=(413, 1), name="sequenceinput"),
    Conv1D(filters=64, kernel_size=3, padding='causal', name="conv1d_1"),
    ReLU(name="relu_1"),
    MaxPooling1D(pool_size=2, padding='same', strides=2, name="maxpool1d_1"),
    LayerNormalization(name="layernorm_1"),
    
    Conv1D(filters=64, kernel_size=3, padding='causal', name="conv1d_2"),
    ReLU(name="relu_2"),
    MaxPooling1D(pool_size=2, padding='same', strides=2, name="maxpool1d_2"),
    LayerNormalization(name="layernorm_2"),
    
    Conv1D(filters=64, kernel_size=3, padding='causal', name="conv1d_3"),
    ReLU(name="relu_3"),
    MaxPooling1D(pool_size=2, padding='same', strides=2, name="maxpool1d_3"),
    LayerNormalization(name="layernorm_3"),
    
    Conv1D(filters=128, kernel_size=3, padding='causal', name="conv1d_4"),
    ReLU(name="relu_4"),
    LayerNormalization(name="layernorm_4"),
    
    GlobalAveragePooling1D(name="globalavgpool1d"),
    Dense(len(np.unique(y_tr_split1)), name="fc"),
    Softmax(name="softmax")
    ])

    X_tr = np.expand_dims(X_tr, axis=-1)
    X_t = np.expand_dims(X_t, axis=-1)
    
    # Compile the model
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    # Train the model
    history = model.fit(X_tr, y_tr, epochs=25, batch_size=32, validation_split=0.2)
    
    # Evaluate the model
    test_loss, test_accuracy = model.evaluate(X_t, y_t)
    print(f'Test accuracy: {test_accuracy:.4f}')
    accuracy_scores.append(test_accuracy)
    
    # Make predictions
    predictions = model.predict(X_t)
    predicted_classes = np.argmax(predictions, axis=1)
    
    # Print a few predictions
    print("Predicted classes:", predicted_classes[:10])
    print("True classes:", y_t[:10])
```
